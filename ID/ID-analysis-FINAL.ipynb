```python
# finding phi at a given angle
k = 2 * np.pi / wavelength
def phi(angle):
    return k * slit_width * np.sin(angle)

# finding the expected amplitude versus angle for each data set
def beta(angle):
    return k * slit_spacing * np.sin(angle)

max_height = np.array([1.49, 1.14, 5.99e-1, 3.53e-1, 2.09e-1])
N = np.array([6.,4.,2.,1.,0.])
A_N = []
for i in range(len(N)):
    if N[i] != 0:  # Avoid division by zero
        A_N.append(max_height[i] / N[i])

# generating angles from -45 to 45 degrees
angles = np.linspace(-.7, .7, 1000)

# zero slit
expected_amplitude_0 = np.zeros_like(angles)

# one slit
A_N_1 = A_N[3]
N_1 = N[3]

expected_amplitude_1 = A_N_1 * np.sin(phi(angles)/2)**2 / (phi(angles)/2)**2 * np.sin(N_1 * beta(angles)/2)**2 / np.sin(beta(angles)/2)**2

# two slits
A_N_2 = A_N[2]
N_2 = N[2]

expected_amplitude_2 = A_N_2 * np.sin(phi(angles)/2)**2 / (phi(angles)/2)**2 * np.sin(N_2 * beta(angles)/2)**2 / np.sin(beta(angles)/2)**2

# four slits
A_N_4 = A_N[1]
N_4 = N[1]

expected_amplitude_4 = A_N_4 * np.sin(phi(angles)/2)**2 / (phi(angles)/2)**2 * np.sin(N_4 * beta(angles)/2)**2 / np.sin(beta(angles)/2)**2

# six slits
A_N_6 = A_N[0]
N_6 = N[0]

expected_amplitude_6 = A_N_6 * np.sin(phi(angles)/2)**2 / (phi(angles)/2)**2 * np.sin(N_6 * beta(angles)/2)**2 / np.sin(beta(angles)/2)**2

# Plot for zero slit
plt.figure(figsize=(10, 6))
plt.plot(t0, a0**2, 'ro', label='Observed Amplitude for 0 slits')
plt.plot(angles, expected_amplitude_0, label='Expected Amplitude for 0 slits')
plt.xlabel('Angle')
plt.ylabel('Amplitude')
plt.title('Amplitude vs Angle for 0 slits')
plt.legend()
plt.grid(True)
plt.show()

# Plot for one slit
plt.figure(figsize=(10, 6))
plt.plot(angles, expected_amplitude_1/20, label='Expected Amplitude for 1 slit')
plt.plot(t1, a1**2, 'ro', label='Observed Amplitude for 1 slit')
plt.xlabel('Angle')
plt.ylabel('Expected Amplitude')
plt.title('Expected Amplitude vs Angle for 1 slit')
plt.legend()
plt.grid(True)
plt.show()

# Plot for two slits
plt.figure(figsize=(10, 6))
plt.plot(angles, expected_amplitude_2/8, label='Expected Amplitude for 2 slits')
plt.plot(t2, a2**2, 'ro', label='Observed Amplitude for 2 slits')
plt.xlabel('Angle')
plt.ylabel('Expected Amplitude')
plt.title('Expected Amplitude vs Angle for 2 slits')
plt.legend()
plt.grid(True)
plt.show()

# Plot for four slits
plt.figure(figsize=(10, 6))
plt.plot(angles, expected_amplitude_4/2, label='Expected Amplitude for 4 slits')
plt.plot(t4, a4**2, 'ro', label='Observed Amplitude for 4 slits')
plt.xlabel('Angle')
plt.ylabel('Expected Amplitude')
plt.title('Expected Amplitude vs Angle for 4 slits')
plt.legend()
plt.grid(True)
plt.show()

# Plot for six slits
plt.figure(figsize=(10, 6))
plt.plot(angles, expected_amplitude_6/1.5, label='Expected Amplitude for 6 slits')
plt.plot(t6, a6**2, 'ro', label='Observed Amplitude for 6 slits')
plt.xlabel('Angle')
plt.ylabel('Expected Amplitude')
plt.title('Expected Amplitude vs Angle for 6 slits')
plt.legend()
plt.grid(True)
plt.show()
```
