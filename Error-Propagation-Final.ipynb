### 1.1 Introduction

You will study error analysis in some detail in the accompanying lectures. Here we give a brief summary of the main results which will enable you to begin making error analyses in an appropriate and acceptable manner. 

We consider an *error* to be the difference between the calculated or measured value of a quantity and the *true* value. Notice that before performing the experiment the true value is unknown. If the true value was known then the estimation of the error would reduce to a trivial subtraction. Of course, if it were known, there would also be no motivation for performing an experiment. In most practical situations we must resort to error analysis to obtain the *uncertainty* in our value. This is an estimate of the expected error between our value and the true value. 

How close your measurements come to the true value is determined by how carefully you control the various sources of error in the experiment. The size of an uncertainty conveys important information about the results. If you quote $t_1 = 10.00 \pm 0.01s$ and $t_2 = 10.0 \pm 0.1s$ then you are telling us that you put more confidence in the estimate of  $t_1$ than $t_2$. These notes are primarily concerned with how we calculate the uncertainty. A careful analysis of the errors in an experiment may also focus attention on the part that most needs improving. Look out for this in your analysis of the experiments in the ENPH 253 laboratory.

### 1.2 Systematic and Random Errors <a id="systematicandrandomerrors"></a>

The process of taking a measurement several times usually generates a spread in the measured parameter. These fluctuations are called *random errors*. A random error varies about the mean value of the quantity being measured and, if you take more and more data, you can make use of statistics to decrease the uncertainty in the final result. If this is the only type of error in an experiment, then in principle there is no limit to this decrease in the uncertainty and your result should converge to the true value given more and more data.  

However, there is another class of experimental error called *systematic error*. Systematic errors may be generated by faulty calibration or poorly designed or constructed apparatus. Maybe your timer runs faster at higher temperatures. A dial might have backlash, so you obtain a different result if you approach the reading from opposite directions. Or a screw might have been poorly cut so that its motion along the screw axis is not a linear function of its rotation as you assumed it to be. In some cases, a systematic error might be constant throughout a set of measurements and, in principle, the data can be corrected later once the source of the error has been identified. This could happen if your voltmeter always reads 1% high. In any event, notice that, unlike the case of random error, repeating the measurements with the same apparatus does not eliminate systematic error, and the final result will generally not converge to the true value as you average more and more data. 

>*Repeated measurements with the same apparatus neither reveal nor do they eliminate a systematic error. For this reason systematic errors are potentially more dangerous than random errors…G L Squires*

In summary, with random errors repeated measurements will give variable results that are equally likely to be higher or lower than the true value. With a systematic error, repeated measurements will give systematically higher or lower values than the true value. 

### 1.3 Accuracy and Precision <a id="accuracyandprecision"></a>

The accuracy of an experiment is a measure of how close the result comes to the true value. The precision of an experiment determines how precisely the result is determined without reference to the accuracy, i.e. how closely do successive determinations agree with each other?

These two parameters are not necessarily related. It is possible to have a very precise experiment that produces results reproducible to 0.1%, but is inaccurate in that the results always differ from the true value by 1%; this would presumably be due to a systematic error in the apparatus. Similarly it is possible to have an accurate, but imprecise, estimate of the true value. In this case repeated measurements might vary by 1% (i.e. the precision is 1%), but on averaging many results the answer might agree with the true value to 0.1%. This will happen only if there are no systematic errors in the apparatus. **A good experiment will simultaneously produce an accurate and precise estimation of the true value.**

### 1.4 Determining Uncertainties From Scales <a id="scales"></a>

For determining uncertainty in measurements with some sort of scale (a ruler, a digital scale, a needle on a dial, etc.) the general rule of thumb is that the error in the measurement is ½ the smallest scale division, **however**, this is only applicable under ideal circumstances. In some cases the uncertainty is greater, for example if you can’t get a ruler right up against what you are trying to measure, if the object has ill-defined edges, or if the digits on a digital display are fluctuating. Often, we can get the best idea of whether our uncertainty is reasonable by repeating the measurement, and we should always do so if we can. If possible have more than one person make the measurement and then look at the range of values. If there are a small number of repeated measurements (2 or 3) half the range of the values should be used as the uncertainty, this also works in the case of a fluctuating digital display. If a larger number of repetitions is made then the standard deviation of the mean can be used as the uncertainty. If repeated measurements are impractical you should consider how confident you are in your ability to read the scale, do you feel that you can read the scale consistently to half the smallest digit, if not to what level of precision?

### 1.5 Propagation of Errors<a id="prop"></a>

#### 1.5.1 Functions of one variable<a id="one-v"></a>

Let $x$ be a measured quantity and $y$ the required quantity defined by $y=f(x)$. This result is valid for both individual measurements and for the mean values. We shall assume that we are dealing with the mean values in what follows. The measurement of $x$ will have a probable error $\Delta x$. Since  

$$\lim_{\Delta x \to 0}\frac{\Delta y}{\Delta x} = \frac{dy}{dx}$$  

the resulting error in $y$ is approximately  

$$\Delta y = \frac{dy}{dx} \Delta x \tag{1}$$  

For example, the area of a circle is given by  

$$A=\pi r^2 \tag{2}$$  

If the radius is measured to be $15.0 \pm 0.1 m$ then an estimation of the area will contain a propagated error $\Delta A$ as follows:  

```python
#Let's find the error in the area numerically:
#First, define your variables.
r = 15.0
Delta_r = 0.1

#Step size for central difference
h = 10e-5

#Next, define Equation 2 as a python function.
def A(r):
    return np.pi*r**2

#Find the derivative of the area with respect to radius. There are many ways to approximate this.
#Here, we used the central difference method.
dAdr = (A(r + h) - A(r - h))/(2*h)

#Calculate the error in the area according to Equation 1.
Delta_A = dAdr*Delta_r

#Print your results.
print('The error in y is:', Delta_A)
print()
print('The final answer is:', round(A(r), 5), '+/-', round(Delta_A, 5), 'm^2')
```

*Note on Significant Figures: The error should be reported to the appropriate number of significant figures, despite the program printing more. This answer would be reported as: $707 \pm 9 \, \text{m}^2$*

*Note: The central difference method is used to calcuate the derivatives and is defined as the following, where $h$ is the step size for the differentiation. In general, a smaller step size results in a smaller error in the approximation.*  

$$f'(x) \approx \frac{f(x+h) - f(x-h)}{2h}$$

#### 1.5.2 Functions of several variables

This can be generalized to the case where you must measure several quantities $a$, $b$, $\dots$ to obtain your final value of a parameter $y$, i.e. $y=f(a,b,\dots)$. In this case we use partial derivatives as follows:  

$$\Delta y=\frac{\partial f}{\partial a}\Delta a+\frac{\partial f}{\partial b}\Delta b \dots$$  

However, if the variables are independent and random, then there is a good chance of partial cancellations of errors (some will be overestimates and some underestimates) and the best estimate of $\Delta y$ is found to be given by  

$$(\Delta y)^2=\left(\frac{\partial f}{\partial a}\Delta a\right)^2+\left(\frac{\partial f}{\partial b}\Delta b\right)^2 \dots \tag{3}$$  

The random errors are said to be added in quadrature when using such relations. 

To demonstrate how you would do this using Python code, let's look at two examples. The first equation we will look at is:

$$y= a + b \tag{4}$$

```python
#To find the error of a function of multiple variables, start by defining your variables.
a = 3.7
b = 4.6
Delta_a = 0.1
Delta_b = 0.2

#Step size for central difference
h = 10e-6

#Next, define Equation 4 as a python function.
def y(a, b):
    return a + b

#Find the partial derivatives with respect to a and b. Again, we used the central difference method.
dyda = (y(a + h, b) - y(a - h, b))/(2*h)
dydb = (y(a, b + h) - y(a, b - h))/(2*h)

#Calculate Delta y according to Equation 3.
Delta_y = np.sqrt((dyda*Delta_a)**2+(dydb*Delta_b)**2)

#Print your results.
print('The error in y is:', Delta_y)
print()
print('The final answer is:', round(y(a, b), 5), '+/-', round(Delta_y, 5))
```

*Note on Significant Figures: The error should be reported to the appropriate number of significant figures, despite the program printing more. This answer would be reported as: $8.3 \pm 0.2$*

Note, you can change the signs on $a$ and $b$ and $\Delta y$ will not change.

The second example we will look at is:

$$y=a^n b^m \tag{5}$$

```python
#Define your variables.
a = 3.7
b = 4.6
n = 2
m = 3
Delta_a = 0.1
Delta_b = 0.2

#Step size for central difference
h = 10e-8

#Define Equation 5 as a python function.
def y(a, b): 
    return a**n * b**m

#Calculate the partial derivatives.
dyda = (y(a + h, b) - y(a - h, b))/(2*h)
dydb = (y(a, b + h) - y(a, b - h))/(2*h)

#Calculate Delta y according to Equation 3.
Delta_y = np.sqrt((dyda*Delta_a)**2+(dydb*Delta_b)**2)

#Print your results.
print('The error in y is:', round(Delta_y, 5))
print()
print('The final answer is:', round(y(a, b), 5), '+/-', round(Delta_y, 5))
```

*Note on Significant Figures: The error should be reported to the appropriate number of significant figures, despite the program printing more. This answer would be reported as: $1300 \pm 200$*

If you had multiple values of $a$ and/or $b$, you could define them as NumPy arrays and then calculate $\Delta y$ the same as before. It's worth noting that not all variables need to be arrays. Say, for example, you have a constant error in $b$, you can define $\Delta b$ as a single value and the calculations will still work.

```python
#Defining a, b, and Delta a as numpy arrays.
a = np.array([3.7, 3.6, 3.5, 4.1, 3.9])
b = np.array([4.6, 4.7, 3.9, 4.2, 4.9])
Delta_a = np.array([0.1, 0.3, 0.4, 0.2, 0.1])

#Define Delta b and h. 
Delta_b = 0.2

#Step size for central difference
h = 10e-6

#Equation 4.
def y(a, b): 
    return a + b

#Find the partial derivatives with respect to a and b using the central difference method.
dyda = (y(a + h, b) - y(a - h, b))/(2*h)
dydb = (y(a, b + h) - y(a, b - h))/(2*h)

#Calculate Delta y according to Equation 3 above.
Delta_y = np.sqrt((dyda*Delta_a)**2+(dydb*Delta_b)**2)

#Print your results.
print('The values of Delta y are:', Delta_y)
```

### 1.6 Sensitivity Analysis

After you have completed an experiment, you may be looking to improve your procedure, possibly by looking at which variable caused the largest error. This can be done though a sensitivity analysis, which looks at each of the variables in the experiment and determines how much each contributes to the total error. 

The formula to calculate the contribution from a variable is related to the equation for adding errors in quadrature. To find the factional error contribution of parameter $a$, then we take the first term in the equation $\left(\frac{\partial f}{\partial a}\Delta a\right)^2$, and divide by the total error $(\Delta y)^2$. Multiplying this by 100 gives the value as a percentage as follows:  

$$\text{% contribution from a} = \left( \frac{\frac{\partial f}{\partial a}\Delta a}{\Delta y} \right)^2\times 100 \% \tag{6}$$  


```python
'''To show how you would do a sensitivity analysis and plot your results, we'll be using the data 
from the first example in Secion 1.5.1.'''

#Here is the data copied over from the previous cell.
a = 3.7
b = 4.6
Delta_a = 0.1
Delta_b = 0.2

#Step size for central difference
h = 10e-6

#Equation 4.
def y(a, b): 
    return a + b

#Find the partial derivatives with respect to a and b using the central difference method.
dyda = (y(a + h, b) - y(a - h, b))/(2*h)
dydb = (y(a, b + h) - y(a, b - h))/(2*h)

#Calculate Delta y according to Equation 3 above.
Delta_y = np.sqrt((dyda*Delta_a)**2+(dydb*Delta_b)**2)

#Find the contribution percentages of a and b using Equation 6.
contrib_a = ((dyda*Delta_a/Delta_y)**2*100)
contrib_b = ((dydb*Delta_b/Delta_y)**2*100)

#Set the figure size
plt.figure(figsize=(10,8))

#Draw the bars
plt.bar(["a", "b"], [contrib_a, contrib_b])
#Argument 1: List of the variable names for the x-axis
#Argument 2: List of the contribution percentages (in the same order as the list of variable names)

#Loop through the list of variables. Enumerate allows variable 'index' to be used as a counter.
for index, data in enumerate([contrib_a, contrib_b]):
    #Add text to the plot
    plt.text(index , data+1 , f"{data:.2f} %", ha='center', color='r', fontsize=15)
    #Argument 1: x axis position of text
    #Argument 2: y axis position of text (1 point higher then the bar for this example)
    #Argument 3: String of text to be plotted
    #Argument 4: Horixontal alignment of text
    #Argument 5: Font size

#Add the figure title and y label
plt.title("Sensitivity Analysis", fontsize=20)
plt.ylabel("Contribution to error [%]", fontsize=15)

#Change size of x and y ticks
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

#Set the y axis to be fixed at the range 0-100
plt.ylim([0, 100])

#Save the figure to the same directory as the notebook
plt.savefig("sensitivityAnalysis.png", dpi=200)
```

### 1.7 Reporting Measurements, Significant Figures

Uncertainties are usually quoted to 1 significant digit and NEVER more than 2 significant digits. Generally, if the first significant digit in the uncertainty is small (say, less than 5) and the second is in the midrange (say 4, 5, 6), then it may make sense to include a second digit since that digit results in an “appreciable change” in the uncertainty. For example, if an uncertainty is determined to be $0.15$, then it might make sense to quote it as $0.15$ instead of $0.2$ since the additional decimal changes the uncertainty by $25\%$ (an arguably appreciable change). Conversely, if the uncertainty is determined to be $0.92$, then it might as well be reported as $0.9$ since the $0.02$ is a small relative change in the uncertainty. 

When reporting results the least significant digit in the quoted value of the measured quantity is determined by the uncertainty, e.g. if the uncertainty is $0.02$ then the result is reported to the nearest hundredth so a value calculated as $2.1386$ would be reported as $2.14 \pm 0.02$. Also, it is much clearer to report the value and the uncertainty using the same exponents so write a result like $2.360\times 10^{-3} \pm 3 \times 10^{-6}$ in the form $(2.360 \pm 0.003) \times 10^{-3}$.

### 2.1 Straight Line Fitting to Data

#### 2.1.1 Method of least squares

In an experiment you may have a set of data which you expect to follow a straight line and you wish to extract the slope and intercept. Although this can be done by eye, it is more conventional and less subjective to use a standard procedure. Methods have been developed that will give the best-fit slope and intercept together with an error estimate to indicate how reliable the parameters are. We will quote without proof the results for the case of a straight line fit to the data using a method based on the minimization of the squares of the differences of the fitted points from the straight line.

The straight line is taken to be of the form  

$$y=A+Bx$$  

and is to be fitted to a data set containing $N$ pairs $(x_i, y_i)$ where the error is assumed to be only in $y_i$, i.e. $x_i$ is arbitrarily accurate. The best estimates of $A$ and $B$ are found to be  

$$A=\frac{\sum{x_i^2}\sum{y_i}-\sum{x_i}\sum{(x_i y_i)}}{\Delta} \tag{7}$$  

and  

$$B=\frac{N\sum{x_i y_i}-\sum{x_i}\sum{y_i}}{\Delta} \tag{8}$$  

where  

$$\Delta = N\sum{x_i^2}- \left(\sum{x_i}\right)^2 \tag{9}$$  

All sums are from $i=1$ to $N$. 

```python
#Create the x and y values.
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 8.11, 9.15, 9.98])

#Define N as the length of the array of x values. 
N = len(x)

#Define the errors.
dx = 0
dy = np.array([0.5, 0.6, 0.4, 0.6, 0.7, 0.5, 0.6, 0.4, 0.6, 0.7])

#Define Delta according to Equation 9.
Delta = N*sum(x**2)-(sum(x))**2

#Define A and B according to Equations 7 and 8.
A = (sum(x**2)*sum(y)-sum(x)*sum(x*y))/Delta
B = (N*sum(x*y)-sum(x)*sum(y))/Delta

#Set the figure size.
plt.figure(figsize=(10,8))

#Plot the points with error bars.
plt.errorbar(x, y, yerr=dy, xerr=dx, fmt='bo', ecolor='k', capsize=3, label="Exp Data")
#Argument 1: x values as a float, list or array
#Argument 2: y values as a float, list or array
#Argument 3: yerr = y error values as a float, list or array
#Argument 4: xerr = x error values as a float, list or array
#Argument 5: fmt = 'colour (b for blue) and marker (o for circles)'
#Argument 6: ecolor = 'colour of the error bars (k for black)'
#Argument 7: capsize = size of the error bar caps
#Argument 8: label = 'label of the line used in the legend

#Plot the stright line fit.
plt.plot(x, A+B*x, 'r', label="Straight line fit")
#Argument 1: x values as a scalar, list or array
#Argument 2: y values as a scalar, list or array
#Argument 3: format the plot, default is a line and 'r' makes it red

#Add the titles, labels, and legend.
plt.legend(fontsize='x-large')
plt.title('Plot with straight line fit', fontsize=20)
plt.xlabel('x axis', fontsize=15)
plt.ylabel('y axis', fontsize=15)

#Add equation of line.
plt.text(x[3], y[6], f"$y={round(B, 2)}x{round(A, 2)}$", color='r', size='x-large')
#Argument 1: x position of text as a scaler, list or array
#Argument 2: y position of text as a scaler, list or array
#Argument 3: text string
#Argument 4: text color
#Argument 5: font size

#Save the figure to the same directory as the notebook.
plt.savefig("plotWithStraightLineFit.png", dpi=200)
```

Often, we wish to force the fitted line to go through the origin, i.e. it might be physically incorrect to have an intercept. In this case we have  

<div align="center">$A=0$ and $B=\frac{\sum{(x_i y_i)}}{\sum x_i^2}$</div>&nbsp;

After obtaining $A$ and $B$, we calculate the deviations $h_i$ of all the measured $y_i$ from the fitted line  

$$h_i=y_i-(A+Bx_i) \tag{10}$$  

We will write the standard deviation of the distribution of the $h$ values as $\sigma_y$; this is defined by  

$$\sigma_y=\sqrt{\frac{\sum{h_i^2}}{N-2}} \tag{11}$$

The *standard error* estimates $\sigma_A$ and $\sigma_B$ in $A$ and $B$ respectively are then given by  

$$\sigma_A=\sigma_y\sqrt{\frac{\sum{x_i^2}}{\Delta}} \tag{12}$$  

and  

$$\sigma_B=\sigma_y\sqrt{\frac{N}{\Delta}} \tag{13}$$  


```python
#Create the x and y values.
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 8.11, 9.15, 9.98])

#Define N as the length of the array of x values. 
N = len(x)

#Define the errors.
dx = 0
dy = np.array([0.5, 0.6, 0.4, 0.6, 0.7, 0.5, 0.6, 0.4, 0.6, 0.7])

#Define Delta according to Equation 9.
Delta = N*sum(x**2)-(sum(x))**2

#Define A and B according to Equations 7 and 8.
A = (sum(x**2)*sum(y)-sum(x)*sum(x*y))/Delta
B = (N*sum(x*y)-sum(x)*sum(y))/Delta


#Define h_i, sigma_y, sigma_A, and sigma_B according to the above Equations 10 through 13.
h_i = y-(A+B*x) #equation 10
sigma_y = np.sqrt(sum(h_i**2)/(N-2)) #Equation 11
sigma_A = sigma_y*np.sqrt(sum(x**2)/Delta) #Equation 12
sigma_B = sigma_y*np.sqrt(N/Delta) #Equation 13

#Print the values.
print('The slope of the line is', round(B, 2), '+/-', round(sigma_B, 2))
print()
print('The intercept of the line is', round(A, 2), '+/-', round(sigma_A, 2))
```

When the line is forced though the origin, the factor $N-2$ in the above equation for $\sigma_y$ is replaced by $N-1$ and the error in the slope is  

$$\sigma_B=\sigma_y\sqrt{\frac{1}{\sum{x_i^2}}}$$  

The results of a least-squares fit are quoted in the form $A\pm \sigma_A$ and $B\pm \sigma_B$. The errors have the normal meaning of a probable error in that there is a $68\%$ probability of finding the true result within the quoted range.

**NOTE: Straight line fitting is done by spreadsheets such as Excel (select *Data Analysis/Regression*) and other software and all use the method just described.**

#### 2.1.2 Weighted Regression

If the uncertainties vary significantly between points *and* you can measure the uncertainties $\sigma_i$ accurately then a weighted fit based on the uncertainty in each point is better than the unweighted fit just described. The rule of thumb for “significantly" is around a factor of 4.

Using the following weighting factor  

$$w_i=\frac{1}{\sigma_i^2} \tag{14}$$  

best estimates of $A$ and $B$ are found to be  

$$A=\frac{\sum{wx^2}\sum{wy}-\sum{wx}\sum{wxy}}{\Delta} \tag{15}$$  

and  

$$B=\frac{\sum{w}\sum{wxy}-\sum{wx}\sum{wy}}{\Delta} \tag{16}$$  

where now  

$$\Delta=\sum{w}\sum{wx^2}-(\sum{wx})^2 \tag{17}$$  

The *standard error* estimates $\sigma_A$ and $\sigma_B$ in $A$ and $B$ respectively are then given by  

$$\sigma_A=\sqrt{\frac{\sum{wx^2}}{\Delta}} \tag{18}$$  

and  

$$\sigma_B=\sqrt{\frac{\sum{w}}{\Delta}} \tag{19}$$  

This algorithm calculates the error in the slope using the measured errors through the inclusion of the weights rather than the variation of the points from the predicted line. If the uncertainties in the measurements are under or overestimated then so will the error in the slope.  

To demonstrate the difference between weighted and unweighted regressions, let's look at a hypothetical experiment. Imagine you have collected the following data in a laboratory experiment and you want to perform a linear regression on it. You have made note of the fact that the uncertainty changes significantly over the data you have collected, and that your data points do not fall perfectly on a linear line as you expected.

```python
#Experimental data
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 9.11, 12.15, 11.68])
dy = np.array([0.17, 0.41, 0.62, 0.79, 1.046, 1.194, 1.37, 1.822, 2.93, 2.836])
dx = 0

#Set the figure size
plt.figure(figsize=(10,8))

#Plot the points with error bars
plt.errorbar(x, y, yerr=dy, xerr=dx, fmt='bo', ecolor='k', capsize=3)

#Add the titles and labels
plt.title("Experimental data", fontsize=20)
plt.xlabel("x (independent variable)", fontsize=15)
plt.ylabel("y (dependent variable)", fontsize=15)

#Draw plot without saving
plt.show()
```

```python
#Unweighted fit
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 9.11, 12.15, 11.68])
dy = np.array([0.17, 0.41, 0.62, 0.79, 1.046, 1.194, 1.37, 1.822, 2.93, 2.836])
dx = 0

#Number of elements
N = len(x)

#Define Delta according to equation 9
Delta = N*sum(x**2)-(sum(x))**2

#Define A and B according to equations 7 and 8
A = (sum(x**2)*sum(y)-sum(x)*sum(x*y))/Delta
B = (N*sum(x*y)-sum(x)*sum(y))/Delta

#Set the figure size
plt.figure(figsize=(10,8))

#Plot the points with error bars
plt.errorbar(x, y, yerr=dy, xerr=dx, fmt='bo', ecolor='k', capsize=3, label="Exp Data")

#Plot the stright line fit
plt.plot(x, A+B*x, 'r', label="Unweighted linear regression")

#Add the titles, labels, and legend
plt.legend(fontsize='x-large')
plt.title("Experimental data with unweighed linear regression", fontsize=20)
plt.xlabel("x (independent variable)", fontsize=15)
plt.ylabel("y (dependent variable)", fontsize=15)

#Add equation of line
plt.text(x[4], y[4]+4, f"$y={round(B, 2)}x{round(A, 1)}$", color='r')

#Draw plot without saving
plt.show()

#Define h_i, sigma_y, sigma_A, and sigma_B according to the above equations 10 through 13
h_i = y-(A+B*x) #equation 10
sigma_y = np.sqrt(sum(h_i**2)/(N-2)) #equation 11
sigma_A = sigma_y*np.sqrt(sum(x**2)/Delta) #equation 12
sigma_B = sigma_y*np.sqrt(N/Delta) #equation 13

#Print the values
print('The slope of the line is', round(B, 2), '+/-', round(sigma_B, 2))
print()
print('The intercept of the line is', round(A, 1), '+/-', round(sigma_A, 1))
```

```python
#Weighted fit
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 9.11, 12.15, 11.68])
dy = np.array([0.17, 0.41, 0.62, 0.79, 1.046, 1.194, 1.37, 1.822, 2.93, 2.836])
dx = 0

#Number of elements
N = len(x)

#Define weights according to equation 14 above
w = 1/dy**2

#Define Delta according to equation 17 above
Delta = sum(w)*sum(w*x**2)-(sum(w*x))**2

#Define A and B using equations 15 and 16 respectively
A = (sum(w*x**2)*sum(w*y)-sum(w*x)*sum(w*x*y))/Delta
B = (sum(w)*sum(w*x*y)-sum(w*x)*sum(w*y))/Delta

#Set the figure size
plt.figure(figsize=(10,8))

#Plot the points with error bars
plt.errorbar(x, y, yerr=dy, xerr=dx, fmt='bo', ecolor='k', capsize=3, label="Exp Data")

#Plot the stright line fit
plt.plot(x, A+B*x, 'r', label="Weighted linear regression")

#Add the titles, labels, and legend
plt.legend(fontsize='x-large')
plt.title("Experimental data with weighted linear regression", fontsize=20)
plt.xlabel("x (independent variable)", fontsize=15)
plt.ylabel("y (dependent variable)", fontsize=15)

#Add equation of line
plt.text(x[4], y[4]+4, f"$y={round(B, 2)}x{round(A, 1)}$", color='r', size='x-large')

#Draw plot without saving
plt.show()

#Define sigma_A, and sigma_B according to the above equations 18 and 19
sigma_A = np.sqrt(sum(w*x**2)/Delta) #equation 18
sigma_B = np.sqrt(sum(w)/Delta) #equation 19

#Print the values
print('The slope of the line is', round(B, 2), '+/-', round(sigma_B, 2))
print()
print('The intercept of the line is', round(A, 1), '+/-', round(sigma_A, 1))
```

<br><br/>
If we qualitatively compare the unweighted and weighted linear regressions, we notice that the weighted regression fits the first 7 data points more accurately (i.e., the data points are closer to the regression line) when compared to the unweighted fit. This is the expected result. Since we set the weights to be $\frac{1}{\sigma^2}$, the larger the uncertainty, the smaller the weight, and the less the regression model will ‘care’ about that data point.

The weighted regression does not fit the last 3 data points as well, however this is acceptable due to the large uncertainty on the measurements. Overall, a weighted regression is preferred for this data since it better represents the data that we are most confident in. In general, if a data set has uncertainties that vary significantly a weighted regression is usually the preferred choice.

### 2.2 Residuals

In order to get an idea of how well our model “fits” the data, one can look at the “residuals” of the fit. For a data point, $(x_i,y_i)$, with a model $y_{model}(x)$, the residuals, $R(x_i)$, are defined as:  

$$R(x_i) \equiv y_i - y_{model}(x_i)$$  

and correspond to the difference between the data and the model prediction. For a linear fit the residuals are equal to the deviations from the straight line that were calculated in the least mean square fit algorithm:  

$$R(x_i) = h_i = y_i - (A+Bx_i) \tag{20}$$  

Again, if we assume that the errors in the $x_i$ are negligible, then the error in the residuals are given by:  

$$\sigma_{R(x_i)} = \sigma_{y_i}$$  

That is, the error on the residuals is given by the error in the data points. If the model is a good representation of the data, then the residuals are expected to be symmetrically distributed about a mean of 0. If all of the residuals were positive or negative, then one would question whether the model is a good fit to the data. Similarly, if the residuals show a trend, then one would also suspect that there is an issue with the model. 

```python
#To show how to make a residuals plot, we'll use the data from section 2.1.1

#Start with the copied data from 2.1.1
#x and y values
x = np.linspace(1, 10, 10)
y = np.array([0.85, 2.05, 3.1, 3.95, 5.23, 5.97, 6.85, 8.11, 9.15, 9.98])

#Define N as the length of the array of x values 
N = len(x)

#Define the errors
dx = 0
dy = np.array([0.5, 0.6, 0.4, 0.6, 0.7, 0.5, 0.6, 0.4, 0.6, 0.7])

#Define Delta according to equation 9
Delta = N*sum(x**2)-(sum(x))**2

#Define A and B according to equations 7 and 8
A = (sum(x**2)*sum(y)-sum(x)*sum(x*y))/Delta
B = (N*sum(x*y)-sum(x)*sum(y))/Delta

#Calculate the residuals according to the above equation
h = y - (A + B*x)

#Set the figure size
plt.figure(figsize=(10,8))

#Plot the points with error bars
plt.errorbar(x, h, yerr=dy, xerr=0, fmt='bo', ecolor='k',capsize=3)
#Argument 1: x values as a float, list or array
#Argument 2: y values as a float, list or array
#Argument 3: yerr = y error values as a float, list or array
#Argument 4: xerr = x error values as a float, list or array
#Argument 5: fmt = 'colour (b for blue) and marker (o for circles)'
#Argument 6: ecolor = 'colour of the error bars (k for black)'
#Argument 7: capsize = size of the error bar caps
#Argument 8: label = 'label of the line used in the legend

#Add the titles and labels
plt.title('Residuals plot', fontsize=20)
plt.xlabel('x points', fontsize=15)
plt.ylabel('distance from the regression line', fontsize=15)

#Save the figure to the same directory as the notebook
plt.savefig("residualsplot.png", dpi=200)
```

### 3.1 Mean, Standard Deviation, and Error on the Mean

When making measurements we would like to eliminate the effect of random errors in measurements, thus revealing the true underlying value of the quantity in question. One method of reducing random error is to repeat the measurement several times. This is done on the assumption that random errors, because they are random, are as likely to make any individual measurement too high as too low. If we make a single measurement we cannot remove the random error because we don’t know whether the measurement is too high or too low. However if we make several measurements then the high errors and low errors will tend to cancel, and the mean value of the measurements will approximate the true value of the quantity. Thus we make repeated measurements in the hope that the (unavoidable) random errors will cancel.

If $N$ measurements of a variable $x_1, x_2, \dots , x_i, x_n$ are obtained, the mean, or average, is defined by  

$$\overline{x} = \frac{1}{N} \sum\limits_{i=1}^{N} x_i \tag{21}$$  

The "sample variance", $\sigma^2_x$, is defined as:  

$$var \equiv \sigma^2_x = \frac{1}{N-1} \sum\limits_{i=1}^{i=N}(x_i - \bar{x})^2 \tag{22}$$  

And the “sample standard deviation”, $\sigma_x$, is defined as the square root of the sample variance:  

$$\sigma_x \equiv \sqrt{var} \tag{23}$$  

While the sample mean gives the average value of the measurement, the sample variance and standard deviation indicate the average distance between the measurements and the sample mean. That is, whether the measured values are all really close to the sample mean, or whether they are spread out. The sample standard deviation is also a representative measure of the uncertainty in a single measurement.

As the number of values $N$ becomes larger and larger, and finally infinite, the sample distribution approaches what is called the *parent distribution*, yielding the parent mean $X$ and the parent standard deviation $\sigma$. It is $X$ that we are trying to measure.

The figure below shows the result of taking 200 measurements of a quantity $x$. The number (or frequency) of the values falling into each ‘bin’ of width 1/3 are given as a function of $x$. (The sum of the numbers in all bins is thus 200). The parent mean is $X$=10.00 and the standard deviation is $\sigma$=1.00. This sample has a mean of $\overline{x}$=9.91 and standard deviation of $\sigma_x$=0.984. The dashed line is a Gaussian using $\overline{x}$ and $\sigma_x$ as the parameters in the equation below.

### 3.2 Probable Error of the Mean

The sample gives a values of $\overline{x}$ and $\sigma_x$ which are estimates of $X$ and $\sigma$. If we took another sample (i.e. another $N$ measurements of $x_i$) we would generally get different values of $\overline{x}$ and $\sigma_x$ as estimates of $X$ and $\sigma$. What then is the best measure of the uncertainty or error between $\overline{x}$ and $X$? At first sight you might think that we just quote $\overline{x} \pm \sigma_x$ *but this is NOT correct*. Clearly $\sigma_x$ is a measure of the error between $x_i$ and $X$ for any single measurement, but we have $N$ measurements so surely the difference between $\overline{x}$ and $X$ is smaller than $\sigma_x$. 

In fact the various values of $\overline{x}$ that we obtain from different samples have their own distribution. This distribution is *narrower* than that for the individual measurements $x_i$ of a sample. To analyze this problem we need to know the details of the parent distribution for $x$. Typically, it is assumed to be *Gaussian*. In this case the probability $P(x)$ of finding a value $x$ in the range $x$ to $x + dx$ is given by  

$$P(x)dx = \frac{1}{\sigma \sqrt{2\pi}} exp \left[- \frac{1}{2} \left(\frac{x-X}{\sigma}\right)^2\right]dx$$  

This has a mean of $X$ and a standard deviation of $\sigma$. If $P(x)$ is integrated over all possible values of $x$, i.e., $-\infty$ to $\infty$, then we obtain unity as expected for probabilities.

It is then found that the distribution for the various sample means $\overline{x}$ is also Gaussian. If $P(\overline{x})$ is the probability of finding a value of $\overline{x}$ in the range $x$ to $x + dx$ then  

$$P(\overline{x}) d\overline{x} = \frac{1}{\sigma_\mu \sqrt{2\pi}} exp \left[- \frac{1}{2} \left(\frac{\overline{x}-X}{\sigma_\mu}\right)^2\right]dx$$  

where $\sigma_\mu = \frac{\sigma}{\sqrt{n}}$. It is the width of this distribution, quantified by $\sigma_\mu$, which is our best estimate of the uncertainty in in our measurements, i.e., the expected difference (the error) between $X$ and $\overline{x}$ for any sample is $\sigma_\mu$. Of course we do not have measurements for the parent distribution so we use our results for the sample distribution to give our best estimates of $X$, $\sigma$ and $\sigma_\mu$. These are  

<div align="center">$X \approx \overline{x}$, $\sigma \approx \sigma_x$, $\sigma_\mu \approx \frac{\sigma_x}{\sqrt{N}}$</div>&nbsp;

**We call $\sigma_\overline{x}$ the probable error of the mean as defined by the following**  

$$\sigma_\overline{x} = \frac{\sigma_x}{\sqrt{N}} \tag{24}$$

**Therefore, the best estimate of X and its uncertainty from our sample is**  

$$x_{best} = \overline{x} \pm \sigma_\overline{x}$$  

As an example, 10 readings of a time interval produce the following values:

$$20.28, 21.26, 20.96, 20.70, 20.31, 21.16, 20.60, 20.36, 20.55, 19.95 \, \text{s}$$  

```python
#Time interval measurements
x = np.array([20.28, 21.26, 20.96, 20.70, 20.31, 21.16, 20.60, 20.36, 20.55, 19.95])

#Number of readings
N = len(x)

#Calculate mean using equation 21
x_bar = 1/N * sum(x)

#Calculate variance using equation 22
var = 1/(N-1) * sum((x - x_bar)**2)

#Calculate standard sample deviation using equation 23
sigma_x = np.sqrt(var)

#Calculate probable error of the mean using equation 24
sigma_x_bar = sigma_x/np.sqrt(N)

print("Mean of x:", round(x_bar, 3), ", Sample Variance:", round(var, 3), ", Sample Standard Deviation:", round(sigma_x, 3), ", Probable Error of the Mean:", round(sigma_x_bar, 3))

print("\n\nHence, x_best = ", round(x_bar, 5), "+/-", round(sigma_x_bar, 5), "\n\n")
```

*Note on Significant Figures: The error should be reported to the appropriate number of significant figures, despite the program printing more. This answer would be reported as: $20.6 \pm 0.1$*

We find $\overline{x} = 20.61$, $\sigma_x = 0.42$, $\sigma_\overline{x} = 0.42/\sqrt{10} = 0.13$, so that $x_{best} = 20.6 \pm 0.1 \, \text{s}$. It is also possible to show that the probability that $\overline{x}$ falls in the range $X \pm \sigma_\overline{x}$ is $0.68$. Likewise, between $X \pm 2\sigma_\overline{x}$ it is $0.95$, and between $X \pm 3\sigma_\overline{x}$ it is $0.997$. These are often called $68\%$, $95\%$, and $99.7\%$ *confidence limits*.  

After completing our measurements and analyzing the data, we usually compare our result with the accepted value of the quantity, if there is one. The accepted value is the best estimate of the true value that you can find in the literature. If you have considered all your sources of error, both systematic and random, your final value should agree with the accepted value within the uncertainties. When doing this, remember that the uncertainties quoted are typically for one standard deviation. If the results do not agree within $3$ standard deviations, i.e. at the $99.7\%$ confidence level, there is probably something badly wrong somewhere.  

**Finally, note that all these error estimates do not include any systematic errors in the apparatus. These must be considered independently.**
